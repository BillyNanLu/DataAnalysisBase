## 1. 折线图

```python
# 绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.sans-serif'] = 'STHeiti'

# 绘图的数据
month = ['1月', '2月', '3月', '4月']
sales = [100, 150, 80, 130]
# 绘制折线图
plt.plot(month, sales,
         label = '产品A',
         color = 'lightpink',
         linewidth = 2,
         linestyle = '--',
         marker = 'o'
         )

# 添加标题
plt.title('2025年销售趋势', fontsize=20)

# 添加坐标轴的标签
plt.xlabel('月份', fontsize=10)
plt.ylabel('销售额（万元）', fontsize=10)

# 添加网格线
plt.grid(True,alpha=0.1, color='blue',linestyle='--')

# 设置坐标轴刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 设置y轴的范围
plt.ylim(0, 200)

# 在每个数据点上显示数值
for x, y in zip(month, sales):
    plt.text(x, y+2, str(y), ha='center', va='bottom', fontsize=10)

plt.show()
```

## 2. 柱状图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'STHeiti'
# 创建图表，设置大小
plt.figure(figsize=(10, 5))

# 绘图的数据
subjects = ['语文', '数学', '英语', '科学']
scores = [85, 92, 78, 88]

# 绘制柱状图
plt.bar(subjects, scores,
        label='Genie',
        color='orange',
        width=0.6)

# 添加标题
plt.title('2025年成绩分布', fontsize=20)
# 添加坐标轴的标签
plt.xlabel('科目', fontsize=10)
plt.ylabel('分数', fontsize=10)
# 添加图例
plt.legend(loc='upper right')
# 添加网格线
plt.grid(axis='y', alpha=0.1, color='blue', linestyle='--')
# 设置刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)
# 设置y轴的范围
plt.ylim(0, 100)
# 在每个数据点上显示数值
for x, y in zip(subjects, scores):
    plt.text(x, y+1, str(y), ha='center', va='bottom', fontsize=10)
# 自动优化排版
plt.tight_layout()
plt.show()
```

## 3. 条形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'STHeiti'

# 创建图表，设置大小
plt.figure(figsize=(10,5))

# 要绘制的锁骨话剧
countries = ['United Stats', 'China', 'Japan', 'Germany', 'India']
gdp = [92, 78, 43, 22, 8]
# 绘制条形图
plt.barh(countries, gdp, color='orange')
# 添加标题
plt.title('2025年GDP排名', fontsize=20)
# 添加坐标轴的标签
plt.xlabel('GDP', fontsize=10)
plt.ylabel('国家', fontsize=10)
# 自动化排版
plt.tight_layout()

plt.show()
```

## 4. 饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'STHeiti'

# 创建图表，设置大小
plt.figure(figsize=(10, 5))
# 绘图的数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
# 绘制饼图
plt.pie(times, labels=things,
        autopct='%.1f%%',
        startangle=90,
        colors=colors
        )
# 添加标题
plt.title('一天的时间分布', color='red', fontsize=20)
# 自动优化排版
plt.tight_layout()
plt.show()
```

## 5. 环形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'STHeiti'

# 创建图表，设置大小
plt.figure(figsize=(10, 5))
# 绘图的数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
# 绘制饼图
plt.pie(times, labels=things,
        autopct='%.1f%%',
        startangle=90,
        colors=colors,
        wedgeprops={'width': 0.7},
        pctdistance=0.6)
# 添加标题
plt.title('一天的时间分布', color='red', fontsize=20)
plt.text(0, 0, '总计：\n100%', ha='center', va='bottom', fontsize=10)
# 自动化排版
plt.tight_layout()
plt.show()
```

## 6. 爆炸式饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'STHeiti'

# 创建图表，设置大小
plt.figure(figsize=(10, 5))
# 绘图的数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
explode = [0.1, 0, 0, 0, 0]
# 绘制饼图
plt.pie(times, labels=things,
        autopct='%.1f%%',
        startangle=0,
        colors=colors,
        explode=explode,
        shadow=True,)
# 添加标题
plt.title('一天的时间分布',fontsize=20)
# 自动化排版
plt.tight_layout()

plt.show()
```

## 8. 箱线图

```python
import matplotlib.pyplot as plt

# 模拟3门课成绩
data = {
    '语文': [82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学': [75, 80, 79, 93, 88, 92, 87, 89, 92],
    '英语': [70, 72, 68, 65, 78, 80, 85, 90, 95]
}
plt.figure(figsize=(8, 6))
plt.boxplot(data.values(), tick_labels=data.keys())

plt.title('各科成绩分布（箱线图）')
plt.ylabel('分数')
plt.grid(True, axis='y', alpha=0.5, linestyle='--')
plt.show()
```

## 9. 多图绘制

```python
import matplotlib.pyplot as plt

# 绘图数据
month = ['1', '2', '3', '4']
sales = [100, 150, 80, 130]

# 动态图表的生成
f1 = plt.subplot(221) # 生成一个子图 行 列 索引
f1.plot(month, sales)
f2 = plt.subplot(222)
f2.bar(month, sales)
f3 = plt.subplot(223)
f3.scatter(month, sales)
f4 = plt.subplot(224)
f4.barh(month, sales)
```

## 1. 温度分析

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'STHeiti'

# 导入数据
df = pd.read_csv('data/weather.csv')
df.head()
# 绘制气温的趋势变化图
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year == 2015]

plt.figure(figsize=(15, 10))

plt.plot(df['date'], df['temp_max'], label='最高气温')
plt.plot(df['date'], df['temp_min'], label='最低气温')

plt.title('2015年气温趋势变化图')

plt.grid(True,alpha=0.1,color='blue',linestyle='--')

plt.xlabel('日期')
plt.ylabel('气温')
plt.legend()
```

```python
# 绘制降水量的直方图
df['temp_mean'] = (df['temp_max'] + df['temp_min']) / 2
plt.figure(figsize=(15, 10))
plt.plot(df['date'], df['temp_mean'], label='平均气温')
plt.title('2015年气温趋势变化图')
plt.xlabel('日期')
plt.ylabel('气温')
plt.legend()
# 绘制降水量的直方图
plt.hist(df['precipitation'], bins=5)
```
